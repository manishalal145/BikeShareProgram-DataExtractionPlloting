```python
# Data Files
vancouver_01_path = os.path.join("resources/Mobi_System_Data_2019-01.csv")
vancouver_02_path = os.path.join("resources/Mobi_System_Data_2019-02.csv")
vancouver_03_path = os.path.join("resources/Mobi_System_Data_2019-03.csv")
vancouver_04_path = os.path.join("resources/Mobi_System_Data_2019-04.csv")
vancouver_05_path = os.path.join("resources/Mobi_System_Data_2019-05.csv")
vancouver_06_path = os.path.join("resources/Mobi_System_Data_2019-06.csv")
vancouver_07_path = os.path.join("resources/Mobi_System_Data_2019-07.csv")
vancouver_08_path = os.path.join("resources/Mobi_System_Data_2019-08.csv")
vancouver_09_path = os.path.join("resources/Mobi_System_Data_2019-09.csv")
vancouver_10_path = os.path.join("resources/Mobi_System_Data_2019-10.csv")
vancouver_11_path = os.path.join("resources/Mobi_System_Data_2019-11.csv")
vancouver_12_path = os.path.join("resources/Mobi_System_Data_2019-12.csv")

# Read Data Files 
vancouver_01 = pd.read_csv(vancouver_01_path, encoding="utf=8")
vancouver_02 = pd.read_csv(vancouver_02_path, encoding="utf=8")
vancouver_03 = pd.read_csv(vancouver_03_path, encoding="utf=8")
vancouver_04 = pd.read_csv(vancouver_04_path, encoding="utf=8")
vancouver_05 = pd.read_csv(vancouver_05_path, encoding="utf=8")
vancouver_06 = pd.read_csv(vancouver_06_path, encoding="utf=8")
vancouver_07 = pd.read_csv(vancouver_07_path, encoding="utf=8")
vancouver_08 = pd.read_csv(vancouver_08_path, encoding="utf=8")
vancouver_09 = pd.read_csv(vancouver_09_path, encoding="utf=8")
vancouver_10 = pd.read_csv(vancouver_10_path, encoding="utf=8")
vancouver_11 = pd.read_csv(vancouver_11_path, encoding="utf=8")
vancouver_12 = pd.read_csv(vancouver_12_path, encoding="utf=8")

# Combine Data into Quarters (Q1. Q2, Q3, Q4)
q1 = pd.concat([vancouver_01, vancouver_02, vancouver_03], axis=1, sort=False)
q2 = pd.concat([vancouver_04, vancouver_05, vancouver_06], axis=1, sort=False)
q3 = pd.concat([vancouver_07, vancouver_08, vancouver_09], axis=1, sort=False)
q4 = pd.concat([vancouver_10, vancouver_11, vancouver_12], axis=1, sort=False)
```
